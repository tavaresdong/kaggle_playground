```python
import sys
import pandas as pd

sys.path.insert(0, "./data")
```

```python
train_data = pd.read_csv("./data/train.csv")
test_data = pd.read_csv("./data/test.csv")
```

```python
all_features = pd.concat((train_data.iloc[:, 1:-1], test_data.iloc[:, 1:]))
```

```python
all_features.shape
```

```python
# Scale the features, and binarize the features
# Q: why do we need the scaler?
import numpy as np
numeric_features = all_features.dtypes[all_features.dtypes != object].index
all_features[numeric_features] = all_features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std()))
all_features = all_features.fillna(all_features.mean())
```

```python
all_features = pd.get_dummies(all_features, dummy_na=True)
all_features.shape
```

```python
all_features.iloc[0:4, [1, 2, 3, 4, -4, -3, -2, -1]]
```

```python
# Turn train and test features into numpy arrays for future use
n_train = train_data.shape[0]
train_features = all_features.iloc[0:n_train, :].values
test_features = all_features.iloc[n_train:, :].values
train_labels = train_data.iloc[:, -1].values
train_labels.shape

all_features.shape
```

```python
# Fit a linear regression model with tensorflow
import tensorflow as tf
import numpy as np

# Placeholders for data input
n_features = train_features.shape[1]
X = tf.placeholder(tf.float32, [None, n_features])
y = tf.placeholder(tf.float32, [None,])
y
```

```python
W = tf.Variable(tf.random_normal(shape=(n_features,1)))
b = tf.Variable(tf.zeros(shape=(1)))
```

```python
y_hat = tf.add(tf.matmul(X, W), b)
y_hat
```

```python
cost = tf.sqrt(tf.reduce_mean(tf.square(tf.subtract(y_hat, y))))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
```

```python
# Train the model for epochs
n_epochs = 10
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(n_epochs):
        c, _ = sess.run([cost, optimizer], feed_dict={X: train_features, y: train_labels})
        print("In epoch " '%04d' %(epoch + 1), " cost is ", '{:.9f}'.format(c))
        
```

```python
# Use scikit-learn
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
train_labels_log = np.log(train_labels)

lin_reg.fit(train_features, train_labels)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error

def score_on_trainset(regressor, features, labels):
    predictions = regressor.predict(features)
    predictions[predictions <= 0] = 1
    mse = mean_squared_log_error(labels, predictions)
    return np.sqrt(mse)

score = score_on_trainset(lin_reg, train_features, train_labels)
score
```

```python
# Use Ridge regression
from sklearn.linear_model import Ridge

ridge_reg = Ridge()
ridge_reg.fit(train_features, train_labels)

ridge_score = score_on_trainset(ridge_reg, train_features, train_labels)
ridge_score
```

```python
# Predict on test data, and write to result file
test_preds = ridge_reg.predict(test_features)
test_preds[test_preds <= 0] = 1
```

```python
test_data["SalePrice"] = test_preds
submission = pd.concat([test_data['Id'], test_data['SalePrice']], axis=1)
submission.to_csv("./submission_linreg.csv", index=False)
```

```python
# Try to use RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=400)
rf_reg.fit(train_features, train_labels)

rf_score = score_on_trainset(rf_reg, train_features, train_labels)
rf_score
```

```python
rf_preds = rf_reg.predict(test_features)
test_data["SalePrice"] = rf_preds
submission = pd.concat([test_data['Id'], test_data['SalePrice']], axis=1)
submission.to_csv("./submission_rf.csv", index=False)
```
